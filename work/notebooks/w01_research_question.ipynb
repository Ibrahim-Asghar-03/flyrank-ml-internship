## 3. Quick look at the data (2-3 real numbers)

Loading the starter dataset validates the scope of this problem. Out of the eligible content inventory, a specific subset is actively showing a downward trend, proving there is a clear imbalance and a concrete workload to prioritize.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

# Assuming the notebook is in work/notebooks/, adjust path if necessary
df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')

# Apply the starter baseline filters to find valid candidates
df_filtered = df[(df['impressions_90d'] > 0) & (df['content_age_days'] >= 90)].drop_duplicates(subset=['content_id'])

eligible_pages = len(df_filtered)
decaying_pages = len(df_filtered[df_filtered['trend_direction'] == 'down'])

print(f"Total eligible candidate pages: {eligible_pages}")
print(f"Pages actively showing a downward trend: {decaying_pages}")
print(f"Percentage of decaying candidates to review: {(decaying_pages / eligible_pages) * 100:.1f}%")
```
